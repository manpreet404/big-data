Q.1 Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('company_sales_data.csv')
data.head()
```

```python
sns.lineplot(x='month_number',y='total_profit',data=data)
plt.title('Q Read Total profit of all months and show it using a line plot')
plt.show()
```

```python
#here just drop unrequire feature
sns.lineplot(data=data.drop(columns=["month_number", "total_units", "total_profit"]),marker='o') # marker='o'
# Set title and labels
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')

# Show plot
plt.legend(title='Product')
plt.show()
```

```python
data_feature=data.drop(columns=['month_number','total_units','total_profit'])
for column in data_feature.columns:
  plt.bar(data['month_number'],data[column],label=column)
plt.legend()
plt.xlabel('month_number')
plt.ylabel('sales')
```

Q.2 Create a Scatter plot in Plotly (can be created using the scatter() method of
plotly.express.)


```python
import plotly.express as px
fig=px.scatter(data,x='month_number',y='total_profit',title='profit over month')
fig.show()
```

Q.3 Create a Bar Chart in Plotly( can be created using the bar() method of
plotly.express class.)

```python
for column in data_feature.columns:
  fig=px.bar(data,x='month_number',y=column)
fig.show()
```

Q.4 Create a histograms in plotly (histograms can be created using the
histogram() function of the plotly.express class.)

```python
fig=px.histogram(data,x="total_units")
fig.show()
```
