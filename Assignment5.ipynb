```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
x=[4,5,10,4,3,11,14,6,10,12]
y=[21,19,24,17,16,25,24,22,21,21]
data = list(zip(x,y))
print(data)
kmeans=KMeans(n_clusters=2)
kmeans.fit(data)
plt.scatter(x,y,c = kmeans.labels_)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Read data from CSV file
df = pd.read_csv('attribute1,attribute2.csv')

# Create a list of tuples containing attribute1 and attribute2 values
data = list(zip(df['attribute1'], df['attribute2']))

# Initialize KMeans with 2 clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Get cluster labels
cluster_labels = kmeans.labels_

# Scatter plot showing the clusters
plt.scatter(df['attribute1'], df['attribute2'], c=cluster_labels)
plt.xlabel('Attribute 1')
plt.ylabel('Attribute 2')
plt.title('KMeans Clustering')
plt.colorbar(label='Cluster Label')
plt.show()
```
