```python
#x=[1,2,3,4,5,6,7]
#y=[9,8,10,12,11,13,14]
x=[2,4,6,8]
y=[3,7,5,10]
n=len(x)
sum_xy=0
sum_xx=0
j=0
for i in x:
  sum_xx+=i*i
  sum_xy+=i*y[j]
  j+=1

b = (n*sum_xy-sum(x)*sum(y))/(n*sum_xx-sum(x)**2)
#a=mean_y-b*mean_x
a=sum(y)/n - b*(sum(x)/n)
#Y=b0+b1*X
print("y =",a," + ",b,"x")
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
Y1=[]
for i in range(n):
    Y = a + b * x[i]
    Y1.append(Y)
    plt.scatter(x[i], y[i],color='blue')  # original data points
    plt.scatter(x[i], Y, color='red')  # points on the regression line
plt.title('SLR')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import numpy as np
x=np.array([[2],[4],[6],[8]])
y=np.array([3,7,5,10])
model =LinearRegression()
model.fit(x,y)
intercept = model.intercept_
slop=model.coef_[0]
y_pred = model.predict(x)

plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='blue')
```
